In order to extract the longest representative path, the following input are required:
* A rGFA file generated by minigraph
* A text file providing weighted value for each nodes 
The weighted value can be obtained by using PARROT (check the Notebook Genome Graph Construction and Analysis).

In this Notebook, Cbc optimizer was used to run the analysis.

```julia
# Load packages
using BioGraph, Cbc, BenchmarkTools
```

```julia
# Read gfa input
@time rgfa = BioGraph.read_from_gfa("platGB2_r551.rgfa", weight_file="platGB2_biograph.txt")
;
```

```julia
# Find graph components
@time graph_component = find_graph_component(rgfa)
;
```

```julia
# Identify the number of simple graph inside the rgfa file
length(graph_component.graph)
```

```julia
# Example when extracting the longest path with 1 simple graph
longest_path_1 = BioGraph.find_longest_path(graph_component.graph[1], Cbc.Optimizer, is_weighted = true)
```

```julia
BioGraph.get_fasta(longest_path_1, header="Graph1", outdir="./")
```

```julia
# Example when extracting the longest path with all simple graph (the number of  loops = the length of graph_component.length)
@time for i in 1:509
    outindex=string("longest_path_", i)
    println(string("Process ",string(outindex)))
    longest_path = BioGraph.find_longest_path(graph_component.graph[i], Cbc.Optimizer, is_weighted = true)
    BioGraph.get_fasta(longest_path, header=outindex, outdir="./")
end
;
```
